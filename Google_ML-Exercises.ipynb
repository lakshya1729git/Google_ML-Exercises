Define plotting functions >

The following code cell uses matplotlib to create the following functions:

plot_the_dataset, which creates a scatter plot of random points in the dataset. Setting the number_of_points_to_plot to 1,400 causes the function to create a scatter plot of all 1,400 points in the dataset.
plot_a_contiguous_portion_of_dataset, which creates a scatter plot of all the points in the dataset from start to end. For example, setting start to 0 and end to 100 will produce a scatter plot of the first 100 points in the dataset.
You may optionally double-click the headline to see the matplotlib code.

```python
# The following code defines the plotting function using matplotlib
def plot_the_dataset(features,labels,number_of_points_to_plot):
  plt.xlabel(features) # days
  plt.ylabel(labels)   # the test score

# Create a scatter plot from n random points of the dataset.
  random_examples = training_df.sample(n=number_of_points_to_plot)
  plt.scatter(random_examples[features], random_examples[labels])

# show the plots / graphs
plt.show()
# Important function
def plot_a_contiguous_portion_of_dataset(features, labels, start, end):

        plt.xlabel(features + 'Day')
        plt.ylabel(labels)
          # Create a scatter plot.
        plt.scatter(training_df[features][start:end], training_df[labels][start:end])
        # Show
        plt.show()


'''When a new plot is created, Matplotlib does not automatically retain labels from previous plots.
If we do not label again, the new plot might have incorrect or missing labels.
Each function is meant to work independently (so if someone calls just this function, the axes are still labeled properly).'''



```

training_df[feature][start:end]

Selects the values from the column feature (x-axis) within the index range start to end.
This extracts a subset of the dataset instead of plotting everything.
training_df[label][start:end]

Selects the values from the column label (y-axis) within the same index range.
These are the corresponding values to be plotted against feature.
plt.scatter(x, y)

Creates a scatter plot using the selected x-values (feature) and y-values (label).

Day 4 must have the problem , all others have max 400 but 200 is the odd one out so 4th day must have the anomaly

```python
# @title Task 5: Solution (expand this code block to view) { display-mode: "form" }

# You could use a variety of metrics to fully compare Thursday to the other
# six days, but this answer simply focuses on the mean.

running_total_of_thursday_calories = 0
running_total_of_non_thursday_calories = 0
count = 0
for week in range(0,4):
  for day in range(0,7):
    for subject in range(0,50):
      position = (week * 350) + (day * 50) + subject
      if (day == 4):  # Thursday
        running_total_of_thursday_calories += training_df['calories'][position]
      else:  # Any day except Thursday
        count += 1
        running_total_of_non_thursday_calories += training_df['calories'][position]

mean_of_thursday_calories = running_total_of_thursday_calories / 200
mean_of_non_thursday_calories = running_total_of_non_thursday_calories / 1200

print("The mean of Thursday calories is %.0f" % (mean_of_thursday_calories))
print("The mean of calories on days other than Thursday is %.0f" % (mean_of_non_thursday_calories))
```
